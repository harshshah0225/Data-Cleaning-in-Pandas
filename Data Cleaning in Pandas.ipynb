```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\iamha\Downloads\Customer Call List.xlsx")
df
```

```python
#remove duplicates
df = df.drop_duplicates()
df
```

```python
#striping off unwanted characters
df["Last_Name"] = df["Last_Name"].str.strip("123./_")
df
```

```python
#reformatting phone numbers into a single format
# df["Phone_Number"] = df["Phone_Number"].str.replace("[^a-zA-Z0-9]", "", regex=True)
# df["Phone_Number"]
# df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
# df["Phone_Number"]
# df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + "-" + x[3:6] + "-" + x[6:10])
df["Phone_Number"] = df["Phone_Number"].replace("nan--", "")
df["Phone_Number"] = df["Phone_Number"].replace("Na--", "")
df
```

```python
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(",",n=2, expand=True)
df
```

```python
df
```

```python
# df["Paying Customer"] = df["Paying Customer"].str.replace("Yes", "Y")
# df["Paying Customer"] = df["Paying Customer"].str.replace("No", "N")
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("Yes", "Y")
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("No", "N")
df
```

```python
df=df.drop(columns="Not_Useful_Column")
```

```python
df
```

```python
df = df.fillna("")
df
```

```python
df=df.drop(columns="Paying Customer\t")
df
```

```python
for i in df.index:
    if df.loc[i, "Do_Not_Contact"] == "Y" or df.loc[i, "Phone_Number"] == "":
        df.drop(i, inplace=True)
df
```

```python
df = df.reset_index(drop=True)
```
